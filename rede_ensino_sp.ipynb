```python
import pandas as pd
```

```python
privada = pd.read_excel("C:/Users/coliv/Documents/Python Scripts/Escolas São Paulo/PRIVADA.XLSX")
```

```python
privada.head()
```

```python
#Selecionando da quarta linha em diante
privada = privada.iloc[4:90]

#Criando uma variável para o novo header com a primeira linha do meu dataframe
new_header = privada.iloc[0] 

#Começando meu dataframe a partir da segunda linha
privada = privada[1:] 

#Nomeando as colunas como a minha variável new_header
privada.columns = new_header
```

```python
#Validando

privada.head()
```

```python
#Eu quero as variáveis de 'Nome dependencia Administrativa' até 'Zona', estou retirando as variáveis de matrículas:

privada = privada.loc[:, 'Nome dependencia Administrativa':'Zona']
```

```python
privada.head() #funcionou.
```
